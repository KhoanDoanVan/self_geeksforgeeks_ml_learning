![](images/image2.png)

# Computation in MLP

- Weights
- Net inputs (sum of weighted inputs)
- Activations (output of neurons to next layer)

# Weights

![](images/image3.png)

$$
W^{(1)} = \begin{bmatrix} w_{11} & w_{12} & w_{13} \\ w_{21} & w_{22} & w_{23} \end{bmatrix}
$$

# Net input

![](images/image4.png)

$$
h^{(2)}=xW^{(1)}=\begin{bmatrix} x_{1} & x_{2} \end{bmatrix} \begin{bmatrix} w_{11} & w_{12} & w_{13} \\ w_{21} & w_{22} & w_{23} \end{bmatrix}
$$

$$
\begin{bmatrix} x_{1}w_{11} + x_{2}w_{21} & x_{1}w_{12} + x_{2}w_{22} & x_{1}w_{13} + x_{2}w_{23}\end{bmatrix} = \begin{bmatrix} h_{1} & h_{2} & h_{3} \end{bmatrix}
$$

# Activation

![](images/image5.png)

$$
a^{(2)} = f(h^{(2)})
$$

# Takeaway points

- ANNs work for complex problems
- Computation is distributed
- Signal moves from left to right
- Weights, net inputs and activations
