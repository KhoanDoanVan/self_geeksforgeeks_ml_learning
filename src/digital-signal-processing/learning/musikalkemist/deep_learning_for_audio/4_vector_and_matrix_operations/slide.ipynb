# Vector

- array of numbers

$$a = \begin{bmatrix} a_{1} & a_{2} & a_{3} \end{bmatrix} = \begin{bmatrix} a_{1} \\ a_{2} \\ a_{3} \end{bmatrix}$$

### Scaler operations

- involve a vector and a number
- addition/substraction/multiplication/division

$$
a + n = \begin{bmatrix} a_{1}\\a_{2}\\a_{3} \end{bmatrix} + n = \begin{bmatrix} a_{1} + n\\a_{2} + n\\a_{3} + n \end{bmatrix}
$$

### Addition/Substraction

- two vectors must have same dimension
- element-wise operation

$$
a + b = \begin{bmatrix} a_{1}\\a_{2}\\a_{3} \end{bmatrix} + \begin{bmatrix} b_{1}\\b_{2}\\b_{3} \end{bmatrix} = \begin{bmatrix} a_{1}+b_{1}\\a_{2}+b_{2}\\a_{3}+b_{3} \end{bmatrix}
$$

$$
a - b = \begin{bmatrix} a_{1}\\a_{2}\\a_{3} \end{bmatrix} - \begin{bmatrix} b_{1}\\b_{2}\\b_{3} \end{bmatrix} = \begin{bmatrix} a_{1}-b_{1}\\a_{2}-b_{2}\\a_{3}-b_{3} \end{bmatrix}
$$

### Dot product

- two vectors involved
- result is a scalar

$$
a.b = \begin{bmatrix} a_{1}\\a_{2}\\a_{3} \end{bmatrix} . \begin{bmatrix} b_{1}\\b_{2}\\b_{3} \end{bmatrix} = a_{1}.b_{1}+a_{2}.b_{2}+a_{3}.b_{3}
$$

# Matrices

- Rectangular grid of numbers (like a spreadsheet)

$$
A_{i,j} = \begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \\ a_{31} & a_{32} \end{bmatrix}
$$

- $i$: row
- $j$: column

### Matrix dimensions

- dimension indicated by rows and columns

### Row and Column Vectors

- Row vector = (1,n) matrix
- Column vector = (n, 1) matrix

### Matrix transposition

- Switch rows and columns

$$
A=\begin{bmatrix} 1 & 2 \\ 3 & -4 \\ 0.5 & 1 \end{bmatrix} \rightarrow A^{T}=\begin{bmatrix} 1 & 3 & 0.5 \\ 2 & -4 & 1 \end{bmatrix}
$$

Inverse:
$$
A^{T}_{ij} = A_{ij}
$$

### Scalar operations

- addition/substraction/multiplication/division of matrix with a number

$$
nA = n \begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \\ a_{31} & a_{32} \end{bmatrix} = \begin{bmatrix} na_{11} & na_{12} \\ na_{21} & na_{22} \\ na_{31} & na_{32} \end{bmatrix}
$$

### Matrix addition/substraction

- matrices must have same dimension
- element-wise operation

$$
\begin{bmatrix} a & b \\ c & d \end{bmatrix} + \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} = \begin{bmatrix} a+1 & b+2 \\ c+3 & d+4 \end{bmatrix}
$$

$$
\begin{bmatrix} a & b \\ c & d \end{bmatrix} - \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} = \begin{bmatrix} a-1 & b-2 \\ c-3 & d-4 \end{bmatrix}
$$

### Matrix multiplication

- of columns of the 1st matrix must be equal of rows of the 2nd
- product of an (m,n) matrix and (n,k) matrix is an (m,k) matrix

$$
\begin{bmatrix} 1 & 2 \\ 3 & 4 \\ 5 & 6 \end{bmatrix} \begin{bmatrix} a & b \\ c & d \end{bmatrix} = \begin{bmatrix} 1a + 2c & 1b + 2d \\ 3a + 4c & 3b + 4d \\ 5a + 6c & 5b + 6d \end{bmatrix}
$$
