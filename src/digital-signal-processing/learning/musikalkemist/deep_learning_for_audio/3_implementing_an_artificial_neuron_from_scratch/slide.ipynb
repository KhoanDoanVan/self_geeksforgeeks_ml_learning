# The Artificial Neuron

## Standard neuron:

$$
z = w^{T}x + b
$$

- $x$: vector input
- $w$: weight of vector
- $b$: bias
- $z$: sum

$$y = σ(z)$$

$σ$ is static and flexible:
- Sigmoid: binary, probability 0-1
- ReLU: hidden layers in deep learning, fast, avoid vanishing
- Tanh: asymmetric nonlinear (-1,1), use in RNN/LSTM
- Softmax: output multi-class
- Leaky ReLU: fixing dead ReLU
- GELU: modern transformer

## Basic architecture of artificial neuron:

![](images/image2.png)

- x: vector
- w: weight of vector x
- $\sum$: sum
- $f$: activation function
- y: output

## Mathematics:

Sum $\sum$:
$$
h = \sum_{i}{x_{i}w_{i}} = x_{1}w_{1} + x_{2}w_{2} + x_{3}w_{3}
$$
Activation $f$:
$$
y = f(h) = f(x_{1}w_{1} + x_{2}w_{2} + x_{3}w_{3})
$$

# The Activation Function: Sigmoid

![](images/image3.png)

## Define:

$\text{Sigmoid}$ is activation function in neural network, used to changes input (maybe positive,negative,infinite,finite) to value between (0,1)

## Mathematics:

Sigmoid:
$$
y = \frac{1}{1 + e^{-x}}
$$

## Apply at Activation Function:

$$
y = \frac{1}{1 + e^{-(x_{1}w_{1}+x_{2}w_{2}+x_{3}w_{3})}}
$$

# Sample

![](images/image4.png)

$h = x_{1}w_{1} + x_{2}w_{2} + x_{3}w_{3}$ = 0.5$*$0.4 + 0.3$*$0.7 + 0.2$*$0.2 = 0.45

$y = \frac{1}{1 + e^{-0.45}}$ = 0.61

# Takeaway points

- Artificial neurons are loosely inspired to biological neurons
- Artificial neurons are computation units
- They transform inputs into outputs using an activation function
