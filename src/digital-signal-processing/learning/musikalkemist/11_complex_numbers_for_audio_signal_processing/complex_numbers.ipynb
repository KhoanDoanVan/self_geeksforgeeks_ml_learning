Calculate $|c|$

$$
 |c|^2 = a^2 + b^2
$$

$$
|c| = \sqrt{a^2 + b^2}
$$

Calculate $\gamma$

$$
cos(\gamma) = \frac{a}{|c|} \space (1)
$$

$$
sin(\gamma) = \frac{b}{|c|} \space (2)
$$

$$
(1),(2)\rightarrow\frac{sin(\gamma)}{cos(\gamma)} = \frac{b}{a} \rightarrow tan(\gamma) = \frac{b}{a}
$$

$$
\gamma = arctan(\frac{b}{a})
$$

Caculate $c$

$$
a = |c|.cos(\gamma) \space (1)
$$

$$
b = |c|.sin(\gamma) \space (2)
$$

$$
c = a + ib \space (3)
$$

$$
c = |c|.(cost(\gamma) + i.sin(\gamma))
$$

# Euler Identity

$Euler's Identity$ is the most special, most beautiful and most famous of $Euler's Formula$

$$
cos(\pi) = -1 \space and \space sin(\pi) = 0
$$
$$
e^{i\pi} = -1 + i0
$$

$$
e^{i\pi} + 1 = 0
$$
