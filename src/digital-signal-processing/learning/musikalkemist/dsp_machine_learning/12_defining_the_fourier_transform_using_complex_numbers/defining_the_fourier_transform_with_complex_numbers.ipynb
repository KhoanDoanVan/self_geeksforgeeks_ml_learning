$$
d_{f} = max_{\phi \in [0,1)}(\int{s(t).sin(2\pi.(ft - \phi)).dt})

# Complex Fourier Transform

$$
\hat{g}(f) = c_{f}
$$

$$
\hat{g}: \mathbb{R} \rightarrow \mathbb{C}
$$

This function take input (frequency: Hert - $\mathbb{R}$) $\rightarrow$ output: fourier transform coefficient (complex number: $\mathbb{C}$)

$$
\hat{g}(f) = \int{g(t).e^{-i2\pi ft}dt}
$$

$$
\hat{g}(f) = \int{g(t).cos(-2\pi ft)dt} + i.\int{g(t).sin(-2\pi ft)dt}
$$

- real part: $\int{g(t).cos(-2\pi ft)dt}$
- imaginary part: $i.\int{g(t).sin(-2\pi ft)dt}$
