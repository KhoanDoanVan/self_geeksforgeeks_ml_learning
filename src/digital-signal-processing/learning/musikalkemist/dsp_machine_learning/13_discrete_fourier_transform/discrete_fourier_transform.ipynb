### Hacking our way around...

$$
\hat{x}(f) = \sum_{n}{x(n).e^{-i2\pi fn}}
$$

The first transform:

$$
\hat{x}(f) = \sum_{n=0}^{N-1}{x(n).e^{-i2\pi fn}}
$$

The second transform:

$$
\hat{x}(\frac{k}{N}) = \sum_{n=0}^{N-1}{x(n).e^{-i2\pi n\frac{k}{N}}}
$$

$$
k = [0, M-1] = [0, N - 1]
$$

# Redundancy in DFT

![](image3.png)

### 1. Where is Redundancy comes from?

- $\text{when you have a real-valued signal}$

- example audio: $$x[n] \in \mathbb{R}$$
- DFT return: $$X[k] \in \mathbb{C}$$
- but: $$$X[k] = X*[N - k]$$

| k   | X[k]   | X[N - k] |
|-----|--------|----------|
| 1   | 3 - 2j | 3 + 2j   |
| 2   | 5 + 1j | 5 - 1j   |
| ... | ... | ...   |

### 2. What's the reason of flex?

- Because if signal as Real, sin signal and cos are create "pair" complex exponential function:
$$cos(\omega t) = \frac{e^{j\omega t} + e^{-j\omega t}}{2}$$

$\rightarrow$ Magnitude must get 2 directions (positive and negative) to descript right phase of signal.

$\rightarrow$ DFT default gets full of both directions, so it will redundan half of signal

### 3. Advantages of Redundancy?

- Reduce computational

$\rightarrow$ Create $\text{Real FFT}$ (rFFT) with half of magnitude

- Reduce memory saving
- Realise real signal more clearly

$\rightarrow$ Negative frequencies exist for mathematics, not for physical
