### Mathematics of STFT

$$
S(m,k) = \sum_{n=0}^{N-1}{x(n + mH).w(n).e^{-i2\pi n\frac{k}{N}}}
$$

- m: frame index
- H: hop length
- w: window function

### From DFT to STFT

DFT mathematics:
$$
\hat{x}(k) = \sum_{n=0}^{N-1}{x(n).e^{-i2\pi n\frac{k}{N}}}
$$
STFT mathematics:
$$
S(m,k) = \sum_{n=0}^{N-1}{x(n + mH).w(n).e^{-i2\pi n\frac{k}{N}}}
$$
Outputs:
- DFT
    - Spectral vector (frequency bins)
    - N complex Fourier coefficients
- STFT
    - Spectral matrix (frequency bins, frames)
    - Complex Fourier coefficients

### Outputs of STFT

number frequency bins:
$$
\frac{framesize}{2} + 1
$$
number frames:
$$
\frac{samples - framesize}{hopsize} + 1
$$

### Example STFT output

- Signal = 10K samples
- Frame size = 1000
- Hop size = 500

- number frequency bins = 1000 / 2 + 1 = 501 -> (0, sampling rate / 2)
- number frames = (10000 - 1000) / 500 + 1 = 19

$$
STFT(501, 19)
$$

### Inputs of STFT

- Frame size
    - 512, 1024, 2048, 4096, 8192
- Hop size
    - 256, 512, 1024, 2048, 4096
- Windowing function

# Visualising sound

$$
Y(m,k) = |S(m,k)|^2
$$

- $S(m,k)$: values of STFT at m frame position and bin of frequency k
- $|S(m,k)|$: magnitude of complex number STFT
- $|S(m,k)|^2$: power
