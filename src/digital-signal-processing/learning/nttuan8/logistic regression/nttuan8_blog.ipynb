# Logistic Regression

Bài trước học về linear regression với đầu ra là giá trị thực,  thì ở bài này sẽ giới thiệu thuật toán logistic regression với giá trị đầu ra là nhị phân. Ví dụ: email gửi đến hòm thư của bạn có phải spam hay không, u là lành tính hay ác tính, ...

### Bài toán

Ngân hàng bạn đang làm có chương trình cho vay ưu đãi cho các đối tượng mua chung cư. Tuy nhiên gần đây có một vài chung cư rất hấp dẫn (giá tốt, vị trí đẹp,…) nên lượng hồ sơ người nộp cho chương trình ưu đãi tăng đáng kể. Bình thường bạn có thể duyệt 10-20 hồ sơ một ngày để quyết định hồ sơ có được cho vay hay không, tuy nhiên gần đây bạn nhận được 1000-2000 hồ sơ mỗi ngày. Bạn không thể xử lý hết hồ sơ và bạn cần có một giải pháp để có thể dự đoán hồ sơ mới là có nên cho vay hay không.

Sau khi phân tích thì bạn nhận thấy là hai yếu tố chính quyết định đến việc được vay tiền đó là mức lương và thời gian công tác. Đây là dữ liệu bạn có từ trước đến nay:

Vậy nên bây giờ bạn không những tìm là hồ sơ ấy cho vay hay không cho vay mà cần tìm **xác xuất** nên cho hồ sơ ấy vay là bao nhiêu.

## Xác suất

Xác suất là gì?

Theo wiki, “Các nhà toán học coi xác suất là các số trong khoảng [0,1], được gán tương ứng với một biến cố mà khả năng xảy ra hoặc không xảy ra là ngẫu nhiên“. Ví dụ bạn tung đồng xu có 2 mặt, thì xác xuất bạn tung được mặt ngửa là 50% ( = 50/100 = 0.5)

Nhận xét:
- Xác xuất của 1 sự kiện trong khoảng [0,1]
- Sự kiện bạn càng chắc chắn xảy ra thì xác suất càng cao. Ví dụ bạn lương cao và còn đi làm lâu năm thì xác suất bạn được vay mua chung cư là cao.
- Tổng xác xuất của sự kiện A và và sự kiện phủ định của A là 100% (hay 1). Ví dụ sự kiện A: tung đồng xu mặt ngửa, xác xuất 50%; phủ định của sự kiện A: tung đồng xu mặt sấp, xác xuất 50% => tổng 100%.

Bạn sẽ thấy xác xuất quan trọng hơn là chỉ 0 hay 1, ví dụ trước mỗi ca mổ khó, bác sĩ không thể chắc chắn là sẽ thất bại hay thành công mà chỉ có thể nói xác xuất thành công là bao nhiêu (ví dụ 80%).

## Hàm Sigmoid

Giờ mình cần tìm xác suất của hồ sơ mới nên cho vay. Hay giá trị của hàm cần trong khoảng [0,1]. Rõ ràng là giá trị của phương trình đường thẳng như bài trước có thể ra ngoài khoảng [0,1]. Nên cần một hàm mới luôn có giá trị trong khoảng [0,1]. Đó là hàm Sigmoid.

![](image4.png)

Đồ thị hàm sigmoid

Nhận xét:
- Hàm số liên tục, nhận giá trị thực trong khoảng (0,1)
- Hàm có đạo hàm tại mọi điểm (để áp dụng gradient descent)

## Model

Với dòng thứ i trong bảng dữ liệu, gọi $x_{1}^{(i)}$ là lương và $x_{2}^{(i)}$ là thời gian làm việc của hồ sơ i.

$p(x^{(i)} = 1) = \hat{y_{i}}$: là xác xuất mà model dự đoán hồ sơ thứ i được cho vay.

$p(x^{(i)} = 0) = 1 - \hat{y_{i}}$: là xác xuất mà model dự đoán hồ sơ thứ i không được cho vay.

$$p(x^{(i)} = 1) + p(x^{(i)} = 0) = 1$$

Sigmoid:
$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Logistic Regression:
$$
\hat{y_{i}} = \sigma(w_{0} + x_{1}^{(i)}.w_{1} + x_{2}^{(i)}.w_{2}) = \frac{1}{1 + e^{-(w_{0} + x_{1}^{(i)}.w_{1} + x_{2}^{(i)}.w_{2})}}
$$

## Loss Function

Giờ cũng cần một hàm để đánh giá độ tốt của model. Như bài trước là $\hat{y}$ càng gần $y$ càng tốt, giờ cũng vậy:

- Nếu hồ sơ thứ i là cho vay, tức $y_{i} = 1$ thì ta cũng mong muốn $\hat{y_{i}}$ càng gần 1 càng tốt hay model dự đoán xác suất người thứ i được vay vốn càng cao càng tốt.
- Nếu hồ sơ thứ i không được vay, tức $y_{i} = 0$ thì ta cũng mong muốn $\hat{y_{i}}$ càng gần 0 càng tốt hay model dự đoán xác suất người thứ i được vay vốn càng thấp càng tốt.

Với mỗi điểm $(x^{(i)}, y_{i})$, gọi hàm loss function:

$$
L = -(y_{i}.log(\hat{y_{i}}) + (1 - y_{i}).log(1 - \hat{y_{i}}))
$$

Thử đánh giá làm $L$ nhé

Nếu $y_{i} = 1$ $\rightarrow$ $L = -log(\hat{y_{i}})$

![](image5.png)

Đồ thị hàm Loss Function trong trường hợp $y_{i} = 1$

- Hàm $L$ giảm dần từ 0 đến 1
- Khi model dự đoán $\hat{y_{i}}$ gần 1, tức giá trị dự đoán gần với giá trị thật $y_{i}$ thì $L$ nhỏ, xấp xỉ 0.
- Khi model dự đoán $\hat{y_{i}}$ gần 0, tức giá trị dự đoán ngược lại giá trị thật $y_{i}$ thì $L$ rất lớn.

$\rightarrow$ Hàm $L$ nhỏ khi giá trị model dự đoán gần với giá trị thật và rất lớn khi model dự đoán sai, hay nói cách khác $L$ càng nhỏ thì model dự đoán càng gần với giá trị thật. $\rightarrow$ Bài toán tìm model trở thành tìm giá trị nhỏ nhất của $L$.

$$
J = -\sum^{N}_{i=1}{(y_{i}.log(\hat{y_{i}}) + (1 - y_{i}).log(1 - \hat{y_{i}}))}
$$

## Chain Rule

Nếu $z = f(y)$ và $y = g(x)$ hay $z = f(g(x))$ thì:

$$
\frac{dz}{dx} = \frac{dz}{dy}.\frac{dy}{dx}
$$

#### Ví dụ:

#### 1. Tính đạo hàm $z(x) = (2x + 1)^{2}$


- Có thể thấy $z = f(g(x))$ trong đó $f(x) = x^{2}$, $g(x) = 2x + 1$. Do đó áp dụng chain rule:

$$\frac{dz}{dx} = \frac{dz}{dy}.\frac{dy}{dx} = \frac{d(2x + 1)^{2}}{d(2x + 1)}.\frac{d(2x + 1)}{dx} = 2.(2x + 1).2 = 4.(2x + 1)$$

- Mọi người biết $\frac{dt^{2}}{dt} = 2t \rightarrow \frac{d(2x + 1)^{2}}{d(2x + 1)} = 2(2x + 1)$ bằng cách đặt $t = 2x + 1$.

Có một cách dễ nhìn hơn để tính chain rule là dùng biểu đồ:

![](image6.png)

- (2,x,1): Số và biến ở bên trái và không có hình tròn bao quanh
- (*,+,^2): Hình tròn là các phép tính

Giá trị của biểu thức sau khi thực hiện phép tính được viết màu đen ở phía tính.
- Ví dụ: phép cộng của 2x và 1 có giá trị 2x+1, phép bình phương 2x+1 có giá trị là (2x+1)^2.

Gía trị của đạo hàm qua thực hiện phép tính được viết ở bên dưới với mực đỏ.
- Ví dụ: phép bình phương, $\frac{d(2x+1)^{2}}{d(2x+1)} = 2(2x + 1)$, hay qua phép cộng $\frac{d(2x+1)}{d(2x)} = 1$, phép nhân $\frac{d(2x)}{d(x)} = 2$

Ở bước cuối cùng đạo hàm được viết là 1 (có hay không cũng được vì nhân với 1 vẫn thế nhưng để 1 để bạn biết là điểm kết thúc).

Giờ càn tính $\frac{d(2x+1)^{2}}{dx}$ thì bạn nhân tất cả các giá trị màu đỏ trên đường đi từ x đến (2x+1)^2

![](image7.png)

Do đó $\frac{d(2x+1)^{2}}{dx} = 2.2(2x+1).1 = 4(2x+1)$

Thực ra nếu bạn để ý biểu đồ chính là chain rule:

$$\frac{d(2x+1)^2}{dx} = \frac{d(2x + 1)^{2}}{d(2x + 1)}.\frac{d(2x + 1)}{d(2x)}.\frac{d(2x)}{dx} = 2.(2x + 1).2 = 4.(2x + 1)$$

**Chỉ là biểu đồ dễ nhìn hơn**

#### 2. Đạo hàm sigmoid $\sigma(x) = \frac{1}{1 + e^{-x}}$

Nhắc lại kiến thức đạo hàm cơ bản:

$$
\frac{d(\frac{1}{x})}{dx} = \frac{-1}{x^2}
$$

$$
\frac{d(e^x)}{dx} = e^x
$$

![](image8.png)

Do đó:
$$
\frac{d(\sigma{(x)})}{dx} = \frac{d(\frac{1}{1 + e^{-x}})}{dx} = \frac{-1}{(1 + e^{-x})^2}.1.e^{-x}.(-1) = \frac{e^{-x}}{(1 + e^{-x})^2} = \frac{1}{1 + e^{-x}}.\frac{e^{-x}}{1+e^{-x}}
$$

$$
= \frac{1}{1+e^{-x}}.(1 - \frac{1}{1 + e^{-x}}) = \sigma(x).(1-\sigma(x))
$$

## Gradient Descent

Với mỗi điểm $(x^{(i)}, y_{i})$, gọi hàm Loss Function:

$$
L = -(y_{i}.log(\hat{y_{i}}) + (1 - y_{i}).log(1 - \hat{y_{i}}))
$$

- trong đó: $\hat{y_{i}} = \sigma(w_{0}+x_{1}^{(i)}.w_{1}+x_{2}^{(i)}.w_{2})$ là giá trị model dự đoán, còn $y_{i}$ là giá trị thật của dữ liệu.

Nhắc lại đạo hàm cơ bản:
$$
\frac{d(log(x))}{dx} = \frac{1}{x}
$$

Áp dụng chain rule:
$$
\frac{dL}{dw_{0}} = \frac{dL}{d\hat{y_{i}}}.\frac{d\hat{y_{i}}}{dw_{0}}
$$

$$
\frac{dL}{d\hat{y_{i}}} = -\frac{d(y_{i}.log(\hat{y_{i}}) + (1-y_{i}).log(1-\hat{y_{i}}))}{d\hat{y_{i}}} = -(\frac{y_{i}}{\hat{y_{i}}} - \frac{1-y_{i}}{(1-\hat{y})})
$$

![](image9.png)

Từ đồ thị ta thấy:

$$
\frac{d\hat{y_{i}}}{dw_{0}} = \frac{\sigma(w_{0}+x_{1}^{(i)}.w_{1}+x_{2}^{(i)}.w_{2})}{dw_{0}} = \hat{y_{i}}.(1-\hat{y_{i}})
$$

$$
\frac{d\hat{y_{i}}}{dw_{1}} = \frac{\sigma(w_{0}+x_{1}^{(i)}.w_{1}+x_{2}^{(i)}.w_{2})}{dw_{1}} = x_{1}^{(i)}.\hat{y_{i}}.(1-\hat{y_{i}})
$$

$$
\frac{d\hat{y_{i}}}{dw_{2}} = \frac{\sigma(w_{0}+x_{1}^{(i)}.w_{1}+x_{2}^{(i)}.w_{2})}{dw_{2}} = x_{2}^{(i)}.\hat{y_{i}}.(1-\hat{y_{i}})
$$

Do đó:

$$\frac{dL}{dw_{0}}=\frac{dL}{dy_{i}}.\frac{dy_{i}}{dw_{0}}=-(\frac{y_{i}}{\hat{y_{i}}}-\frac{(1-y_{i})}{(1-\hat{y_{i}})}).\hat{y_{i}}.(1-\hat{y_{i}})=-(y_{i}.(1-\hat{h_{i}})-(1-y_{i}).\hat{y_{i}})=\hat{y_{i}}-y_{i}$$

Tương tự:

$$
\frac{dL}{dw_{1}}=x_{1}^{(i)}.(\hat{y_{i}}-y_{i})
$$

$$
\frac{dL}{dw_{2}}=x_{2}^{(i)}.(\hat{y_{i}}-y_{i})
$$


Đây là trên một điểm dữ liệu, trên toàn bộ dữ liệu:

$$
\frac{dL}{dw_{0}}=\sum^{N}_{i=1}(\hat{y_{i}}-y_{i})
$$

$$
\frac{dL}{dw_{1}}=\sum^{N}_{i=1}{x_{1}^{(i)}.(\hat{y_{i}}-y_{i})}
$$

$$
\frac{dL}{dw_{2}}=\sum^{N}_{i=1}{x_{2}^{(i)}.(\hat{y_{i}}-y_{i})}
$$
Biểu diễn bài toán dưới dạng ma trận:

$$
X = \begin{bmatrix} 1&x_{1}^{(1)}&x_{2}^{(1)}\\1&x_{1}^{(2)}&x_{2}^{(2)}\\1&...&...\\1&x_{1}^{(n)}&x_{2}^{(n)} \end{bmatrix},
y = \begin{bmatrix} y_{1}\\y_{2}\\...\\y_{n} \end{bmatrix},
w = \begin{bmatrix} w_{0}\\w_{1}\\w_{2} \end{bmatrix}
$$

$$
\hat{y} = \sigma(Xw)
$$

$$
J = -\sum(y.log(\hat{y})+(1-y).log(1-\hat{y}))
$$

$$
\frac{dJ}{dw} = X^{T}.(\hat{y}-y) \space , \space X^{T}=\begin{bmatrix} 1&1&...&1\\x_{1}^{(1)}&x_{1}^{(2)}&...&x_{1}^{(n)}\\x_{2}^{(1)}&x_{2}^{(2)}&...&x_{2}^{(n)} \end{bmatrix}
$$
- Lý do **Toán Học** nhân với Transpose của X là:
    - $X(N\times d) \rightarrow X^{T}(d\times N)$
    - $(\hat{y}-y)$ là $(N\times 1)$

- Lý do **Thật sự** nhân với Transpose của X là:
    - Bởi vì gradient theo từng weight $w_{j}$ phải có dạng: $\sum_{i=1}^{N}{(\hat{y_{i}}-y_{i})x_{ij}}$
        - Đây là **tổng theo từng cột** của X (mỗi cột là một feature)

Sau khi thực hiện thuật toán gradient descent, ta tìm được $w_{0},w_{1},w_{2}$. Với mỗi hồ sơ mới $x^{(t)}$ ta sẽ tính được phần trăm nên cho $\hat{y_{t}} = \sigma(w_{0} + x_{1}^{(t)}.w_{1} + x_{2}^{(t)}.w_{2})$ rồi so sánh với ngưỡng cho vay của công ty t (bình thường là t = 0.5, thời kì thiết chặt thì là t = 0.8), nếu $\hat{y_{t}} \geq t$ thì cho vay, không thì không cho vay.

## Quan hệ giữa phần trăm và đường thẳng.
Xét đường thẳng y = ax+b, gọi f = y – (ax+b) thì đường thẳng chia mặt phẳng thành 2 miền, 1 miền hàm f có giá trị dương, 1 miền hàm f có giá trị âm và giá trị f các điểm trên đường thẳng bằng 0
![](image10.png)

Ví dụ, xét đường thẳng y = 3-x, hàm f = y – 3 + x.
- Tại điểm A(1,3) => f = 3 – 3 + 1 = 1 > 0
- Tại điểm B(2,1) nằm trên đường thẳng => f = 2 – 3 + 1 = 0
- Tại điểm C(1,1) => f = 1 – 3 + 1 = -1 < 0

Giả sử bạn lấy mốc ở chính giữa là 50% tức là nếu hồ sơ mới dự đoán $\hat{y_{i}}$ >= 0.5 thì cho vay, còn nhỏ hơn 0.5 thì không cho vay.

![](image11.png)

$\rightarrow$ đường thẳng $w_{0}+w_{1}.x+w_{2}.y=0$ chỉnh là đường phân cáhc giữa các điểm cho vay và từ chối.

![](image12.png)

Trong trường hợp tổng quát bạn lấy xác suất lớn hơn $t$ (0<t<1) thì mới cho vay tiền

$\hat{y_{i}}>t$ <=> $w_{0} + x_{1}^{(i)}.w_{1} + x_{2}^{(i)}.w_{2}$ > $-\ln(\frac{1}{t} - 1)$

![](image13.png)


Ví dụ t = 0.8, bạn thấy đường phân chia gần với điểm màu đỏ hơn so với t = 0.5 thậm chí một số hồ sơ cũ trước được cho vay nhưng nếu giờ nộp lại cũng từ chối. Nghĩa là công ty thắt chặt việc cho vay lại. Nghe hoàn toàn hợp lý đúng không?
