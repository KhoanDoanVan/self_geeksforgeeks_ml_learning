```python
import matplotlib.pyplot as plt
from IPython.core.pylabtools import figsize
from skimage.feature import hog
from skimage import data, exposure
```

```python
image = data.coffee()

image
```

Compute HOG features

```python
fd, hog_image = hog(image, orientations=8, pixels_per_cell=(16, 16), cells_per_block=(1, 1), visualize=True, channel_axis=-1)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4), sharex=True, sharey=True)
```

```python
fig.suptitle('Coffee Cup Image with Histogram of Oriented Gradients (HOG)', fontsize=14, fontweight='bold')
```

```python
ax1.imshow(image, cmap=plt.cm.gray)
```

```python
ax1.set_title('Original Image')
```

```python
ax1.axis('off')
```

Rescale histogram for better display

```python
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
```

HOG Features

```python
ax2.imshow(hog_image_rescaled, cmap=plt.cm.gray)
```

```python
ax2.set_title('HOG Features')
```

```python
ax2.axis('off')
```

```python
plt.tight_layout()
plt.subplots_adjust(top=0.85) # Adjust the layout to accommodate the title
plt.show()
```
