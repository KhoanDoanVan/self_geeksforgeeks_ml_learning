```python
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import sklearn.datasets
```

```python
california = sklearn.datasets.fetch_california_housing()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    california['data'],
    california['target'],
    test_size=0.33,
    random_state=42
)
```

```python
# Data scaling
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Squared Error

```python
clf = SGDClassifier(
    loss='squared_error',
    max_iter=10000
)
```

```python
clf.fit(X_train, y_train)
```

```python
y_test_pred_squarederror = clf.predict(X_test)
```

```python
print('\033[1m' + 'Squared Error' + '\033[0m')
print('Coefficient of determination: ', metrics.r2_score(y_test, y_test_pred_squarederror))
```

Squared Error
Coefficient of determination:  0.9128269523492567

# Huber Error

```python
clf_huber = SGDClassifier(
    loss='huber',
    max_iter=10000
)
```

```python
clf_huber.fit(X_train, y_train)
```

```python
y_test_pred_huber = clf_huber.predict(X_test)
```

```python
print('\033[1m' + 'Huber Error' + '\033[0m')
print('Coefficient of determination: ', metrics.r2_score(y_test, y_test_pred_huber))
```

Huber Error
Coefficient of determination:  0.9148961830574946

# Epsilon Insensitive

```python
clf_epsilon = SGDClassifier(
    loss='epsilon_insensitive',
    epsilon=1,
    max_iter=10000
)
```

```python
clf_epsilon.fit(X_train, y_train)
```

```python
y_test_pred_epsilon = clf_epsilon.predict(X_test)
```

Epsilon Insensitive Loss Function
Coefficient of determination:  0.4266448634462471
